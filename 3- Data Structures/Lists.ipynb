```python
## Accessing the list items

fruits = ["apple", "banana", "cherry", "Kiwi", "guava"]
print(fruits[0])
print(fruits[3])
print(fruits[-1])

## if you want to print all the items only for index 1 to n:

print(fruits[1:])
print(fruits[1:3])
```

```python
## Modify the list elements

fruits
```

```python
fruits[1] = "watermelon"
print(fruits)
```

```python
## List Methods

fruits.append("orange") ## it will add an item at the end 
print(fruits)

## if you wan to add at specific index

fruits.insert(1, "watemelon")
print(fruits)

## To remove

fruits.remove("watermelon")
print(fruits)
```

```python
print(fruits)
fruits.remove("orange") ## remove the first occurence of an item

```

```python
## remove and return the last occurance of an item
popped_fruits= fruits.pop()
print(popped_fruits)
print(fruits)

```

```python
idx = fruits.index("cherry")
print(idx)
```

```python
fruits.insert(2, "banana")
print(fruits.count("banana"))
print(fruits)
```

```python
fruits.sort()
print(fruits)
```

```python
fruits.reverse() ## reverse the list
print(fruits)
```

```python
fruits.clear() ## remove all the from the list
print(fruits)
```
