## Logical Operators

```python
## AND, NOT, OR
a = 10
b = 5

print(a > 5 and b > 3)   # True
print(a > 5 and b > 10)  # False

print(a > 5 or b > 10)   # True
print(a < 5 or b > 10)  # False

is_logged_in = False

print(not is_logged_in)  # True
```
