```python
# for Loop

for i in range(5):
    print(i)
```

```python
## While loop

## The while lopp countinues to execute as long as the condition is true.

count = 0

while count < 5:
    print(count)
    count = count + 1
```

```python
## Loop Control Statements

## Break Statement

for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
## Continue Statement
for i in range(10):
    if i%2 ==0:
        continue
    print(i)
 
```

```python
## Pass

for i in range(5):
    if i == 3:
        pass
    print(i)
```

```python
## Nested Loops

for i in range(3):
    for j in range(2):
        print(f" {i} and {j} ")
    
```

```python
## Example: Calculte sum of n natural number

## While Loop

n = 10
sum = 0
count  = 1

while count <= n:
    sum = sum + count
    count = count + 1 

print(sum)
```

```python
## for Loop

n = 10
sum = 0


for i in range(11):
    sum += i

print(sum)
```

```python
## Prime Number

for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                break
        else:
            print(num)
```
